### Python Chatbot Implementation

The previous cell `GG5AKyZ0hixc` contained Java code, which resulted in a `SyntaxError` because Google Colab cells execute Python by default. Below is a Python implementation that replicates the logic of your Java `getResponse` method for a text-based chatbot.

```python
def get_response(message):
    message = message.lower()

    if "hello" in message or "hi" in message:
        return "Hello! How can I help you?"
    elif "how are you" in message:
        return "I am fine. Thanks for asking!"
    elif "your name" in message:
        return "My name is AI ChatBot."
    elif "java" in message:
        return "Java is a popular programming language."
    elif "ai" in message:
        return "AI stands for Artificial Intelligence."
    elif "machine learning" in message:
        return "Machine Learning helps systems learn from data."
    elif "bye" in message:
        return "Goodbye! Have a nice day."
    else:
        return "Sorry, I don't understand."
```

```python
print("AI ChatBot Started (Python version)")

while True:
    user_input = input("You: ")
    response = get_response(user_input)
    print(f"Bot: {response}")

    if user_input.lower() == "bye":
        break
```
